### Create data and import matplotlib 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*3
z = x**3
```

### 1. Plot (x,y) on a figure on a canvas with width and height of 1, and location at (0,0), label the axis and name it figure Figure 1

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green')
ax.set_title('Figure 1')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

### 2. Create a inset figure 

Create a same inset figure use (x,z) (ax2) inside in the above figure at location of [0.2, 0.5, 0.2, 0.2]

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax.plot(x,y, color='green')
ax.set_title('Figure 1')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax2.plot(x,z, color='orange')
ax2.set_title('Inset Figure 2')
ax2.set_xlabel('x')
ax2.set_ylabel('y') 
```

### 3. Create a inset (width=0.3, height=0.3) Figure that zoom Figure 1 from x:20-22, y:30-50

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

ax.plot(x, y, color='green')
ax.set_title('Figure 1')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax2.plot(x, y,color='blue')
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim((20, 22))
ax2.set_ylim((50, 80))
```

### 4. Plot 2 figures side by side on one canvas 

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, 'orange', lw=2, ls='--')
ax[0].set_title('Figure 1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x, z, 'blue', lw=2)
ax[1].set_title('Figure 2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
plt.tight_layout()
```

**Change figure size**

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

ax[0].plot(x, y, 'orange', lw=2, ls='--')
ax[0].set_title('Figure 1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x, z, 'blue', lw=2)
ax[1].set_title('Figure 2')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')

plt.tight_layout()
```
